```python
import os
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib.colors import Normalize
import re  # Import the regular expression module
import pandas as pd

# Directory containing your text files
data_directory = "/Users/arpi/data/02.02.24/processed"

background = pd.read_csv('/Users/arpi/data/30.01.24/processed/processed_CAF9.txt', header = None, delimiter=' ')

norm_tran = background[1]

# Get a list of all text files in the directory
file_list = [file for file in os.listdir(data_directory) if file.endswith(".txt")]

# Initialize a plot
plt.figure(figsize=(10, 6))

# Iterate over each file in the directory
for file in file_list:
    # Create the full path to the file
    full_path = os.path.join(data_directory, file)

    # Read the file using pd.read_csv
    data = pd.read_csv(full_path, header=None, delimiter=' ')

    # Divide the second column of the file by norm_tran
    data[1] = data[1] / norm_tran

    # Plot the data
    plt.scatter(background[0], data[1], label=file, s = 2)

# Set plot labels and legend
plt.xlabel('First Column of Background')
plt.ylabel('Y-axis (Modified Second Column)')
plt.legend()

# Show the plot
plt.show()



```
