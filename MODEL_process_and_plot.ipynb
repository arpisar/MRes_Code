```python
import os
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm

data_path = '/Users/arpi/Desktop/MRes/Measurements/09.01.24'


def process_files(input_directory):
    for filename in os.listdir(input_directory):
        if filename.endswith(".txt"):
            input_file_path = os.path.join(input_directory, filename)

            # Read data from the input file
            with open(input_file_path, 'r') as file:
                lines = file.readlines()

            # Process data and create new columns
            output_lines = []
            for line in lines:
                wavenumber, transmission = map(float, line.strip().split())

                # Convert wavenumber to wavelength
                wavelength = 1 / wavenumber * 1e4

                # Create new column names
                wavelength_col_name = f"wavelength_{filename}"
                transmission_col_name = f"transmission_{filename}"

                # Append results to the output lines
                output_lines.append(f"{wavelength} {transmission / 100}\n")

            # Write the processed data to a new file on a new folder called 'processed'
            output_file = os.path.join(f'{input_directory}/processed/processed', f"processed_{filename}")
            with open(output_file, 'w') as file:
                file.writelines(output_lines)

            print(f"Processing complete for {filename}. Results saved to {output_file}")




process_files(data_path)


```

```python
# Directory containing your text files that you want to plot

file_path = (f'{data_path}/processed/processed')

# Function to plot the data from a file 
def plot_file(file_path):
    data = np.loadtxt(file_path)
    wavelength = data[:, 0]
    transmission = data[:, 1]

    plt.xlabel('Wavelength (um)')
    plt.ylabel('Transmission Percentage')
    plt.title('Transmission vs Wavelength')
    plt.legend()
    


    plt.scatter(wavelength, transmission, s=2)

plt.show()




# Get a list of all text files in the directory
file_list = [os.path.join(file_path, file) for file in os.listdir(file_path) if file.endswith(".txt")]


for file in file_list:
    plot_file(file)








```
