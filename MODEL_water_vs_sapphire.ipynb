```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
import seaborn as sns
```

```python
# Importing values
water = pd.read_csv('/Users/arpi/data/water_transmission/water_tranmsission.csv', header=None)
sapphire_cold = pd.read_csv('/Users/arpi/data/10.10.23/processed/processed_Al2031.txt', delimiter=' ', header=None)
sapphire_roomtemp = pd.read_csv('/Users/arpi/data/10.10.23/processed/processed_Al20318.txt', delimiter=' ', header=None)
gas = pd.read_csv('/Users/arpi/data/water_transmission/water_gas_transmission.csv', header=None)


water_tran = water[1]
water_wave = water[0]

gas_tran = gas[1]
gas_wave = gas[0]


sapphire_ctran = sapphire_cold[1]
sapphire_cwave = sapphire_cold[0]

sapphire_htran = sapphire_roomtemp[1]
sapphire_hwave = sapphire_roomtemp[0]


```

```python
plt.scatter(water_wave,water_tran, s = 2, label = 'Water in liquid form')
plt.scatter(sapphire_cwave,sapphire_ctran, s =2, label = 'Sapphire 77K')
plt.scatter(sapphire_hwave,sapphire_htran, s =2, label = 'Sapphire room temperature')
plt.plot(gas_wave,gas_tran, color = 'red', label = 'Water in gas form')
plt.legend()

plt.savefig("Water_vs_Sapphire.png", dpi=500)


plt.show()

sapphire_coldp = pd.read_csv('/Users/arpi/data/10.10.23/processed/neartest_n_averaging/nearest_n_avg_processed_Al2031.txt',delimiter=' ')
sapphire_roomtempp = pd.read_csv('/Users/arpi/data/10.10.23/processed/neartest_n_averaging/nearest_n_avg_processed_Al20318.txt', delimiter=' ')

x1 =sapphire_coldp['Wavelength']
y1 = sapphire_coldp['Smoothed_Transmission']

x2 = sapphire_roomtempp['Wavelength']
y2 = sapphire_roomtempp['Smoothed_Transmission']


plt.scatter(water_wave,water_tran, s = 2, label = 'Water in liquid form')
plt.scatter(x1, y1, s =2, label = 'Sapphire 77K - smoothed')
plt.scatter(x2, y2, s =2, label = 'Sapphire room temperature - smoothed')
plt.plot(gas_wave,gas_tran, color = 'red', label = 'Water in gas form')
#plt.legend()
plt.savefig("Water_vs_Sapphire_smoothed.png", dpi=500)
plt.show()
```
